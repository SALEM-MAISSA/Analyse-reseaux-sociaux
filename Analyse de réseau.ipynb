```python


import pandas as pd
import networkx as nx

```

```python

us_graph=nx.from_pandas_edgelist(df,source='Source',target='Target')
```

```python


type(us_graph)

```

```python
# Graph Density
nx.density(us_graph)
```

```python
# Show node degree
nx.degree(us_graph)
```

```python


# Visualization

import matplotlib.pyplot as plt 
%matplotlib inline 

```

```python


nx.draw(us_graph)

```

```python


plt.figure(figsize=(10,10))
nx.draw_networkx(us_graph)
plt.show()

```

```python


# Most Influential
nx.degree_centrality(us_graph)

```

```python


sorted(nx.degree_centrality(us_graph))

```

```python
most_inf=nx.degree_centrality(us_graph)
```

```python
for w in sorted(most_inf, key=most_inf.get,reverse=True):
    print(w, most_inf[w])
```

```python
# Most important Connection 
nx.eigenvector_centrality(us_graph)
```

```python
# Betweener Centrality 

plt.figure(figsize=(10,10))
nx.draw_networkx(us_graph)
plt.show()
```

```python


nx.betweenness_centrality(us_graph)

```
